```python
import os
import json
os.chdir("C:/users/nicol/master_thesis/code")
from src.eval.postprocess import binary_fill_tumor,remove_small_blobs_v2,smoothen_prediction, remove_small_dhgp_blobs
from imageio.v3 import imread
from PIL import Image
```

```python
EXPERIMENT_FOLDER_NAME = "final_ssl_dice_extended"
splits_config = json.load(open("./src/config/splits_faber.json"))
new_folder = f"D:/tfm_data/hpc_results/ablationstudy_results/{EXPERIMENT_FOLDER_NAME}_pp_edited"
os.makedirs(new_folder,exist_ok=True )
for split in [1]:
    print("starting split", split)
    split_config = splits_config[f"split_{split}"]
    labeled_ids = split_config["test_wsi_ids"]

    split_dir = os.path.join("D:/tfm_data/hpc_results/ablationstudy_results", EXPERIMENT_FOLDER_NAME, f"split{split}")
    labeled_split_folder = os.path.join(split_dir, "labeled")
    
    new_split_folder = os.path.join(new_folder, f"split{split}")
    os.makedirs(new_split_folder,exist_ok=True )

    for labeled_id in labeled_ids:
        # Load the final ensemble
        print("starting id", labeled_id)
        ensemble_prediction = imread(os.path.join(labeled_split_folder, f"{labeled_id}_finalensemble.png"))
        smoothened_pred_array = smoothen_prediction(ensemble_prediction, disk_size=20)
        blobs_removed = remove_small_blobs_v2(smoothened_pred_array,min_blob_size=8000)
        dhgp_islands_removed = remove_small_dhgp_blobs(blobs_removed)
        filled_tumors = binary_fill_tumor(dhgp_islands_removed)
        Image.fromarray(filled_tumors).save(os.path.join(new_split_folder, f"{labeled_id}_postprocessed.png"))
```
